```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [16.0,8.0]
```

```python
from google.colab import drive
df1 = pd.read_excel('/content/6291002-2.xlsx','Data2')

df2 =df1.drop(df1.index[0:9])
```

```python
df2.rename(columns={'Unnamed: 0': 'Date'}, inplace=True)
df2["Date"] = pd.to_datetime(df2["Date"])
df2['Year'] = df2['Date'].dt.year
df2 = pd.concat([df2.iloc[:, 87:168], df2.iloc[:, -1]], axis=1)

df2.head(50)
```

```python
grouped_by_year = df2.groupby("Year")[[
    ">> Greater Brisbane ;  > Employed full-time ;  > Females ;",
    ">> Greater Brisbane ;  > Employed full-time ;  > Males ;",
    ">> Greater Brisbane ;  > Employed part-time ;  > Males ;",
    ">> Greater Brisbane ;  > Employed part-time ;  > Females ;",
]].sum()
grouped_by_year.head(50)
```

```python
# Columns to clean
columns_to_convert = [
    ">> Greater Brisbane ;  > Employed part-time ;  > Males ;",
    ">> Greater Brisbane ;  > Employed part-time ;  > Females ;",
    ">> Greater Brisbane ;  > Employed full-time ;  > Males ;",
    ">> Greater Brisbane ;  > Employed full-time ;  > Females ;",
]

for col in columns_to_convert:
    grouped_by_year[col] = pd.to_numeric(
        grouped_by_year[col].astype(str).str.replace(",", "").str.strip(),  # Remove commas and spaces
        errors="coerce"  # Convert invalid entries to NaN
    )
```

```python
import plotly.express as px

# Plot the data with Year as the x-axis
fig = px.line(
    grouped_by_year.reset_index(),  # Temporarily reset the index to access "Year"
    x="Year",
    y=[
        ">> Greater Brisbane ;  > Employed full-time ;  > Males ;",
        ">> Greater Brisbane ;  > Employed full-time ;  > Females ;",
        ">> Greater Brisbane ;  > Employed part-time ;  > Males ;",
        ">> Greater Brisbane ;  > Employed part-time ;  > Females ;",
    ],
    title="Employment Trends by Year in Greater Brisbane",
    labels={"value": "Number of Employees", "Year": "Year", "variable": "Category"},
)

fig.update_layout(
    xaxis_title="Year",
    yaxis_title="Number of Employees",
    legend_title="Categories",
    hovermode="x unified",
)

fig.show()
```

### Employment Trends in Greater Brisbane

This graph shows that both male and female full-time workforces in Greater Brisbane have been increasing over time. However, a key insight is that the gap between male and female full-time employment remains relatively constant throughout the entire period, indicating a persistent disparity in workforce participation.

This graph suggests that a significant portion of women in Brisbane hold part-time jobs. This trend could be attributed to factors such as maternity responsibilities or a preference for dedicating part of their time to household activities and caregiving duties. Furthermore, the combined number of women employed in both full-time and part-time roles appears to be similar to the total number of men employed full-time.

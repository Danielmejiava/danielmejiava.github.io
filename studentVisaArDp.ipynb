```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [16.0,8.0]
```

```python
from google.colab import drive
arrivals = pd.read_excel('/content/3401015.xlsx',sheet_name="Table 15.1",header=16)
departures = pd.read_excel('/content/3401015.xlsx', sheet_name="Table 15.2",header=16)

arrivals['Month'] = pd.to_datetime(arrivals['Month'], errors='coerce')
arrivals.rename(columns={'Month': 'Date'}, inplace=True)

departures['Month'] = pd.to_datetime(departures['Month'], errors='coerce')
departures.rename(columns={'Month': 'Date'}, inplace=True)


arrivals['Year'] = arrivals['Date'].dt.year.astype('Int64')
departures['Year'] = departures['Date'].dt.year.astype('Int64')

arrivals.info()
departures.info()
```

```python
arGroupYear= arrivals.groupby('Year')[[
    'Temporary Student Visas',
    'Total(h)']].sum()

deGroupYear= departures.groupby('Year')[[
    'Temporary Student Visas',
    'Total(h)']].sum()
```

```python
arGroupYear.rename(columns={
    'Temporary Student Visas' : 'Temporary Student Visas(Arrivals)',
    'Total(h)': 'Total(h)(Arrivals)'
},inplace=True)

deGroupYear.rename(columns={
    'Temporary Student Visas' : 'Temporary Student Visas(Departures)',
    'Total(h)': 'Total(h)(Departures)'
},inplace=True)
```

```python
# Merge both dataframes on the index (Year)
merged_df = arGroupYear.merge(deGroupYear, left_index=True, right_index=True)
```

```python
merged_df.head()
```

```python
merged_df['% Participation Arrivals'] = (merged_df['Temporary Student Visas(Arrivals)'] / merged_df['Total(h)(Arrivals)']) * 100

merged_df['% Participation Departures'] = (merged_df['Temporary Student Visas(Departures)'] / merged_df['Total(h)(Departures)']) * 100
```

```python
merged_df.head(21)
```

```python
import plotly.graph_objects as go

# Create an interactive line plot for Temporary Student Visas (Arrivals & Departures)
fig = go.Figure()

# Add Temporary Student Visas (Arrivals) as a line chart
fig.add_trace(go.Scatter(
    x=merged_df.index,
    y=merged_df['Temporary Student Visas(Arrivals)'],
    mode='lines+markers',
    name='Temporary Student Visas (Arrivals)',
    line=dict(color='blue')
))

# Add Temporary Student Visas (Departures) as a line chart
fig.add_trace(go.Scatter(
    x=merged_df.index,
    y=merged_df['Temporary Student Visas(Departures)'],
    mode='lines+markers',
    name='Temporary Student Visas (Departures)',
    line=dict(color='red', dash='dash')
))

# Set layout for the timeline chart
fig.update_layout(
    title="Timeline of Temporary Student Visas (Arrivals & Departures)",
    xaxis_title="Year",
    yaxis_title="Number of Visas",
    template="plotly_dark",
    hovermode="x unified"
)

# Show the interactive line chart
fig.show()
```

```python
import plotly.graph_objects as go

# Create an interactive bar chart for % Participation Arrivals vs % Participation Departures
fig = go.Figure()

# Add % Participation Arrivals as a bar chart
fig.add_trace(go.Bar(
    x=merged_df.index,
    y=merged_df['% Participation Arrivals'],
    name='% Participation Arrivals',
    marker_color='green'
))

# Add % Participation Departures as a bar chart
fig.add_trace(go.Bar(
    x=merged_df.index,
    y=merged_df['% Participation Departures'],
    name='% Participation Departures',
    marker_color='orange'
))

# Set layout for the bar chart
fig.update_layout(
    title="% Participation Arrivals vs % Participation Departures",
    xaxis_title="Year",
    yaxis_title="% Participation",
    barmode='group',  # Group bars side by side
    template="plotly_dark",
    hovermode="x unified"
)

# Show the interactive bar chart
fig.show()
```

- Post-2020, **both arrivals and departures recover significantly**.

The graph highlights the **impact of global events** on student mobility, showcasing:
